```python
df1 = pd.DataFrame({'key': np.repeat(1, ndf.shape[0]), 'UserID': ndf.UserID})
df2 = pd.DataFrame({'key': np.repeat(1, ndf1.shape[0]), 'ProductID': ndf1.ProductID})
user_products = pd.merge(df1, df2,on='key')[['UserID', 'ProductID']]
pd.merge(user_products, ndf, how='left', on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(lambda x: pd.Series(dict(
    Quantity=x.Quantity.sum()
))).reset_index().fillna(0)
```
